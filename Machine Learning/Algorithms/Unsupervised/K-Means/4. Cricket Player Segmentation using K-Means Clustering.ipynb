```python
from sklearn.cluster import KMeans
```

```python
k_value=[2,3,4,5,6,7,8,9]
```

```python
wcss=[]
for k in k_value:
  #we are iterating 100 times since Kmeans requires calculating centroid and clustering repetitively,
  #so more iterations would make sure that the clusters are perfect
  model=KMeans(n_clusters=k,max_iter=100,random_state=23) #max_iter: maximum iteration
  model.fit(df_copy_scaled)
  wcss.append(model.inertia_) #inertia_:data points are spread far apart from their assigned cluster centroids
```

```python
wcss
```

```python
plt.plot(k_value, wcss)#optimal k is chosen where we can find a clear difference(bend) in the graph
```

```python
#Silhoutte Score
from sklearn.metrics import silhouette_score
num_clusters = [i for i in range(2, 10)]
def kmeans_sil(num_clusters, X):

    sil_score = []
    for k in num_clusters:
        kms = KMeans(n_clusters=k)
        kms.fit(X)
        sil_score.append(silhouette_score(X, kms.labels_))

    return sil_score
```

```python
sil_score = kmeans_sil(num_clusters, df_copy)
sil_score
```

```python
# Create a line plot of silhouette scores
plot = sns.lineplot(x=num_clusters, y=sil_score)
plot.set_xlabel("# of clusters");
plot.set_ylabel("Silhouette Score");
```
