```python
from sklearn.preprocessing import StandardScaler
```

```python
df_copy=df.copy()
```

```python
df_copy.head()
```

```python
scalar = StandardScaler()
```

```python
df_scaled = scalar.fit_transform(df_copy)
```

```python
df_scaled
```

```python
df_scaled = pd.DataFrame(df_scaled,columns=df_copy.columns)
```

```python
df_scaled
```

```python

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
#Elbow method
wcss=[]
for k in range(2,10):
  model=KMeans(n_clusters=k,max_iter=100,random_state=50)
  model.fit(df_scaled)
  wcss.append(model.inertia_)
```

```python
wcss
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(2,10),wcss,marker='o')
plt.xlabel("No of Clusters - K value")
plt.ylabel("WCSS")
plt.title("Elbow method")
plt.show()
```

####After k=2 there is a fast drop. and after k=3 drop curve is less sharp.  So here K = 3

```python
#Silhouette score

sil_score=[]
for k in range(2,10):
  kms=KMeans(n_clusters=k,random_state=50)
  labels = kms.fit_predict(df_scaled)
  #kms.fit(df_scaled)
  score = silhouette_score(df_scaled,labels)
  sil_score.append(score)
```

```python
sil_score
```

```python
plt.figure()
plt.plot(range(2,10), sil_score, marker='o')
plt.title("Silhouette Score")
plt.xlabel("Number of Clusters - K value")
plt.ylabel("Score")
plt.show()
```

### As per graph above, for k=3 is the maximum value so besk K value is 3

```python
final_model=KMeans(n_clusters=3, max_iter=100, random_state=40)
```

```python
final_model.fit(df_scaled)
```

```python
final_model.labels_
```

```python
df['ClusterID']=final_model.labels_
```

```python
df
```
