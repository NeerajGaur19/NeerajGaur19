#KNN (K-Nearest Neighbours)
- it predicts a label for a new data point by looking at its neighbors.

**STEPS**
1. Pick a K value (K=number of neighbors)
2. For a new data point(called as query point), calculate its distance(Euclidean distance) from all other points in the dataset.
3. Find the K closest points(neighbors)
4. Classify on the basis of majority

```python
df=pd.read_csv('/content/dataset.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#finding the percentange of null values in each column
#if a column has more than 50% null values, drop that column

df.isnull().mean()*100
```

```python
df_new=df[['Disease','Symptom_1','Symptom_2','Symptom_3','Symptom_4','Symptom_5','Symptom_6']]
```

```python
df_new
```

```python
df_new.isnull().sum()
```

```python
for col in df_new.columns:
  if (df_new[col].dtype)=='object':
    df_new[col]=df_new[col].fillna(df_new[col].mode()[0])
  else:
    df_new[col]=df_new[col].fillna(df_new[col].mean())
```

```python
df_new['Symptom_1'].mode()[0]
```

```python
df_new.isnull().sum()
```

#Why not remove duplicates in KNN?
- In KNN, every data points(row) is like a vote.
- If two data points are exactly the same(duplicates), they both vote for the same class.
- That's not wrong -> actually make the classification stronger.
- Removing duplicates would weaken the evidence

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

for col in df_new.columns:
  df_new[col]=le.fit_transform(df_new[col])

  #enumerate: gives both index and category name, therefore we are using 2 variable
  for idx,cls in enumerate(le.classes_):
    print(f"{cls} -> {idx}")
```

```python
#----CLUSTERING ALGO----
```

```python
x=df_new.drop('Disease', axis=1)
y=df_new['Disease']
```

```python
x
```

```python
y
```

```python
y.unique()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#To pick best K for KNN
#THUMB RULE: set k as square root of total number of n
#Elbow method
#Cross Validation (testing multiple k values)

#BEST PRACTICE
#- Use Odd numbers
#- Consider Dataset Noise:noisy dataset-> larger k, and vice versa
#- Complexity vs Simplicity:
  #  - Small k(eg: k=1): High complexity, low bias but high variance
  #  It captures fine details but often overfits by reacting to noise
  # - Large k: Low complexity, low variance, but high bias.
  #  It produces smoother decision boundaries but may miss local patterns(underfitting)
```

```python
from sklearn.neighbors import KNeighborsClassifier

#by default: KNN= 5 K + Euclidean Distance + Voting; n_neighbors=k
model=KNeighborsClassifier()

model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import *

acc=accuracy_score(y_pred,y_test)
#TESTING ACCURACY
acc
```

```python
# .score() -> predicts and gives accuracy
#TRAINING ACCURACY
acc2=model.score(x_train,y_train)
acc2
```
