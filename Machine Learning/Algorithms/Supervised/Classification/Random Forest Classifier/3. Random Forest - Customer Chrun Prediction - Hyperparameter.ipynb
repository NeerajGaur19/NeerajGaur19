```python
# most of the columns are having data type -> object
# convert categorical columns data to numerical data ------ Label Encoding
# 12:08 p.m
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()

# create a list with all object data type column so that you can convert to numercal
col_list = []
for i in data.columns:
  if((data[i].dtype=='object') & (i!='Churn')):
    col_list.append(i)

col_list
```

```python
for i in col_list:
  data[i] = le.fit_transform(data[i])
```

```python
# divide the data into x and y-

x = data.iloc[:,:-1]
y = data['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=0 , stratify=y )
```

```python
from random import Random
from sklearn.ensemble import RandomForestClassifier
```

```python
# HyperParameters -> settings/controls that you choose before training the model
# they are not learned from data
# n_estimators -> number of trees
# max_depth -> maximum depth of each tree
# min_samples_split -> minimum samples needed to split the data
# min_samples_leaf -> minimum samples in leaf
# max_features -> number of features to consider at each split
```

```python
model1 = RandomForestClassifier(n_estimators= 100)   # number of decision trees that you want to use = 50
```

```python
model1.fit(x_train,y_train)
```

```python
ans = model1.predict(x_test)
```

```python
ans
```

```python
# 12:46
from sklearn.metrics import *
```

```python
accuracy_score(ans,y_test)
```

```python
cm = confusion_matrix(ans,y_test)
cm
```

```python
# use different values for n_estimaror
n = [50,60,100,200,300,400,500,600,700,800,900,1000]

accuracy = []  # in this list i am going to store all the accuracy
for i in n:
  model = RandomForestClassifier(n_estimators = i)
  model.fit(x_train,y_train)
  ans = model.predict(x_test)
  a = accuracy_score(y_test,ans)*100
  accuracy.append(a)
```

```python
accuracy
```

```python
d = {'number of trees':n , 'accuracy of them in RF':accuracy}
df = pd.DataFrame(d)
df
```

```python
# how to choose the optimal value for our hyperparameters
# HYPERPARAMETER TUNNING - choosing the best hyperparameter setting to get the highest accuracy/best performance
# 1- Grid Search CV(cross-validation) -> try all possible combinations
# cv -> how many folds you want -> cv = 5 (equal folds/parts) -> model train 4 parts , model test 1 parts
```

```python
# Grid Search CV

from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators':[100,200,300],
    'max_depth':[None,5,10],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,3],
    'criterion':['gini','entropy']
}
```

```python
model2 = RandomForestClassifier(random_state=0)
# peform grid search cv
grid_search = GridSearchCV(estimator=model2,param_grid = param_grid , cv=2)
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_estimator_
```

```python
# you can build a model using all these values for hyperparameter
```

```python
model3 = RandomForestClassifier(n_estimators= 200,max_depth=10,min_samples_leaf=2)
model3.fit(x_train,y_train)
ans = model3.predict(x_test)
```
