```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/census-income.csv")
df
```

```python
df['workclass'].unique()
```

```python
df.isnull().sum()
```

```python
df.replace("?",np.nan,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

LE=LabelEncoder()

for col in df.columns:
  if df[col].dtype=='object':
    df[col]=LE.fit_transform(df[col])
```

```python
df
```

```python
x=df.drop('annual_income', axis=1)
y=df['annual_income']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

#Adaboost

```python
from sklearn.ensemble import AdaBoostClassifier

ada_model=AdaBoostClassifier()

ada_model.fit(x_train,y_train)

y_pred=ada_model.predict(x_test)
```

```python
from sklearn.metrics import *

acc=accuracy_score(y_test,y_pred)
print(acc)
```

#Gradient Boost

```python
#In classification, GB uses LogLoss(Loss function) to  evaluate how well predicted probabilities align with the class labels(0 or 1)
#and penalizes confident, incorrect predictions
#Using LR and the same steps for regression, it reduces the loss function
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_model=GradientBoostingClassifier()

gb_model.fit(x_train,y_train)

y_pred=gb_model.predict(x_test)
```

```python
from sklearn.metrics import *

acc=accuracy_score(y_test,y_pred)
print(acc)
```

#XGBoost

```python
#Tianqui Chen
#XGBOOST focuses on: 1. performance,
#2. speed(larger and different datasets)
#3. flexibility(compatible with different languages, OS and libraries)
```

```python
#XGBoost uses: 1. Regularization: adding penalties to reduce overfitting by preventing the model from becoming too complex
#2. Parallelization: evaluates possible splits in parallel across features and then picks the best
#3. Tree Pruning: (Pre-pruning: Gradient Boosting), Post-Pruning, saves pre processing time
```

```python
#L1(Lasso): adds the sum of the "ABSOLUTE" values of the coefficients(weights) as a penalty to the loss function
#Sets many feature weights to zero because of which automatically feature selection happens

#Add absolute value of weights to the loss function -> forces less important columns to become exactly zero
```

```python
#L2(Ridge): adds the sum of the squared values of the coefficients(weights) as a penalty to the loss function
#Shrink all weights uniformly towards zero, but rarely sets them to exact zero.
```

```python
#Elastic Net: hybrid of L1 and L2
```

```python
#Gradient Boost - General physician, XGBoost - Cardiologist
```

```python
from xgboost import XGBClassifier

xgb_model=XGBClassifier()

xgb_model.fit(x_train,y_train)

y_pred=xgb_model.predict(x_test)
```

```python
from sklearn.metrics import *

acc=accuracy_score(y_test,y_pred)
print(acc)
```
