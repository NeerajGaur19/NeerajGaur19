# Text to Audio

```python
from transformers import pipeline
```

```python
generator = pipeline(
    "text-to-audio",
    model="facebook/musicgen-small"
)

audio = generator("Happy piano music")
```

```python
print(audio)
```

```python
from scipy.io.wavfile import write

write(
    "relaxing_music.wav",
    rate=audio["sampling_rate"],
    data=audio["audio"]
)
```
